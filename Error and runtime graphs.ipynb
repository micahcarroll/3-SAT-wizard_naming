```python
import runner
import numpy as np
import matplotlib.pyplot as plt
```

# 3SAT

## Runtime and errors with respect to number of wizards

```python
wizards = np.arange(10, 50, 1);
time = np.vectorize(lambda x: runner.run(x, 100)[0]);
errors = np.vectorize(lambda x: runner.run(x, 100)[1]);
plt.plot(wizards, time(wizards), wizards, errors(wizards));
plt.show()
```

## Runtime and errors with respect to number of constraints (20 wizards)
